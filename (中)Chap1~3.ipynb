<br>sqrt(s(s-a)(s-b)(s-c))

```python
import math
a,b,c=3,4,5
s=(a+b+c)/2
d=math.sqrt(s*(s-a)*(s-b)*(s-c))
print(d)
```

### 雞兔同籠 若共有35個頭 94隻腳 請問有幾隻雞?幾隻兔

```python
for chicken in range(0,36):
    rabbit=35-chicken
    if(2*chicken+4*rabbit)==94:
        print("雞有",chicken,"隻","兔子有",rabbit,"隻")
        break
else:
    print("無解")
```

# Tuple(元祖)
tuple(元祖)可以視為不可改變的串列(list)。tuple跟list很像，但是不能新增刪除或著更新tuple的元素。
tuple比起list的好處：
* 占用空間少
* 可以當作dictionary(字典)的key
* 具名tuple當作物件代替
* 當作函數的引數

```python
a_list=['python','js','C#']
a_tuple=tuple(a_list)
b_tuple=('python','js','C#')
print(a_tuple)
print(b_tuple)
print(a_tuple[1])
```
